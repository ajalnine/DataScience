```python
def getfriends(userid):
    url = "https://api.vk.com/method/"
    method="friends.get"
    r = requests.post(url + method,
                      data={"user_id":str(userid),
                            "v":"5.69"
                      })

    j = json.loads(r.text)
    return j["response"]["items"]

def getuser(userid):
    url = "https://api.vk.com/method/"
    method="users.get"
    r = requests.post(url + method,
                      data={"user_id":str(userid),
                            "fields" : "bdate, crop_photo",
                            "v":"5.69"
                      })

    j = json.loads(r.text)
    t = j["response"][0]
    return t

def sendmessage(userid, text):
    url = "https://api.vk.com/method/"
    method="messages.send"
    r = requests.post(url + method,
                      data={"user_id":str(userid),
                            "message" : text,
                            "access_token" : access_token,
                            "v":"5.69"
                      })

    j = json.loads(r.text)
    print(j)
    return j

def recievelastmessage(userid):
    url = "https://api.vk.com/method/"
    method="messages.get"
    r = requests.post(url + method,
                      data={"access_token" : access_token,
                            "v":"5.69"
                      })

    j = json.loads(r.text)
    print(j)
    return [x["body"] for x in j["response"]["items"] if x["user_id"]==userid][0]
```

```python
for a in getfriends(3215191):
    t = getuser(a)
    date = None
    if ("bdate" in t.keys()):
        date = "(" + t["bdate"] + ")"
    print ("{} {} {} {}".format(t["first_name"], t["last_name"], date, t["id"]))
    if ("crop_photo" in t.keys()):
        display(Image(t["crop_photo"]["photo"]["photo_604"], width=302, unconfined=True))
```
