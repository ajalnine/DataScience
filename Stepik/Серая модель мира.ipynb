```python
%matplotlib inline
from skimage.io import imread, imsave, imshow
from skimage import img_as_float, img_as_ubyte
import numpy as np

rgb = img_as_float(imread('img.png'))
av = rgb.mean(axis=(0,1))
rgb *= av.mean() / av

imsave("out_img.png", img_as_ubyte(rgb.clip(0,1)))
imshow("out_img.png")
```
