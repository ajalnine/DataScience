```python
from llama_cpp import Llama

# Путь к локальной GGUF модели
model_path = "./qwen3-8b-Q8_0.gguf"

# Загрузка модели (только CPU)
llm = Llama(
    model_path=model_path,
    n_ctx=2048,  # Максимальная длина контекста
    n_threads=8,  # Количество потоков (настройте под ваш CPU)
    n_gpu_layers=-1,  # 0 для CPU-only
    verbose=False  # Отключение логов
)

# Функция для генерации текста
def generate_text(prompt, max_tokens=100):
    response = llm(
        prompt,
        max_tokens=max_tokens,
        temperature=0.8,  # Креативность
        top_p=0.95,  # Контроль разнообразия
        echo=False  # Не включать промпт в ответ
    )
    return response["choices"][0]["text"].strip()

# Пример использования
prompt = "Напиши песню в стиле Есенина про времена года, объем песни - 5 куплетов с припевами, формат для Suno"
response = generate_text(prompt, max_tokens=10000)
print("Ответ модели:", response)

# Очистка памяти
del llm
```
